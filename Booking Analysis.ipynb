```python
perc_cancel=df['is_canceled'].value_counts(normalize=True)
print(perc_cancel)
plt.title('Reservation Status Count')
bars=plt.bar(['Not_canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.bar_label(bars)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is the original DataFrame
# Filtering data for resort and city hotels
resort_hotels = df[df['hotel'] == 'Resort Hotel']
city_hotels = df[df['hotel'] == 'City Hotel']

# Grouping and calculating the mean ADR
resort_hotels_grouped = resort_hotels.groupby('reservation_status_date')[['adr']].mean()
city_hotels_grouped = city_hotels.groupby('reservation_status_date')[['adr']].mean()

# Plotting both in one plot
fig, ax = plt.subplots(figsize=(15, 8))

resort_hotels_grouped.plot(color='teal', title='ADR Comparison: Resort Hotels vs. City Hotels', ax=ax)
city_hotels_grouped.plot(color='coral', ax=ax)

# Adding labels and legend
ax.set_xlabel('Reservation Status Date')
ax.set_ylabel('Average Daily Rate (ADR)')
ax.legend(['Resort Hotels ADR', 'City Hotels ADR'])

plt.tight_layout()
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(20,8))
ax1=sns.countplot(x='month',hue='is_canceled',palette='bright',data=df)
plt.title('Reservation Status per month',size=20)
for container in ax1.containers:
    ax1.bar_label(container)
plt.xlabel('Month')
plt.ylabel('Number of cancellation')
plt.legend(['Not Cancelled','Cancelled'])
plt.tight_layout
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame and it has a 'month' column

# Grouping data by month and summing the 'adr' for canceled reservations
monthly_adr = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
print(monthly_adr)
# Creating the bar plot
plt.figure(figsize=(20,8))
plt.title('Adr Per month',fontsize=20)
ax=sns.barplot(x='month', y='adr', data=monthly_adr,color='#FF5733')
for container in ax.containers:
    ax.bar_label(container)
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Filter data for canceled reservations
c_d = df[df['is_canceled'] == 1]
c_d_adr = c_d.groupby('reservation_status_date')[['adr']].mean().reset_index()
c_d_adr = c_d_adr[(c_d_adr['reservation_status_date'] >= '2016-01-01') & (c_d_adr['reservation_status_date'] <= '2017-12-31')]
c_d_adr.sort_values('reservation_status_date', inplace=True)

# Filter data for non-canceled reservations
n_d = df[df['is_canceled'] == 0]
n_d_adr = n_d.groupby('reservation_status_date')[['adr']].mean().reset_index()
n_d_adr = n_d_adr[(n_d_adr['reservation_status_date'] >= '2016-01-01') & (n_d_adr['reservation_status_date'] <= '2017-12-31')]
n_d_adr.sort_values('reservation_status_date', inplace=True)

# Plotting the data
plt.figure(figsize=(14, 7))
plt.plot(c_d_adr['reservation_status_date'], c_d_adr['adr'], label='Canceled Reservations', color='teal')
plt.plot(n_d_adr['reservation_status_date'], n_d_adr['adr'], label='Non-Canceled Reservations', color='coral')
plt.title('Average Daily Rate (ADR) from 2016 to 2017')
plt.xlabel('Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
